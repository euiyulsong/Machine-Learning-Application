```python
def meanquadraticloss(lm, x, y):
    return np.square(y - lm.predict(x)).mean()
x = pd.DataFrame(bdata.data, columns = bdata.feature_names)
y = bdata.target
arr = pd.DataFrame([], columns = ['Alpha', 'Ridge MSE 10CV'])
i = 0
for l in lam:
    lm = Ridge(fit_intercept=True, alpha=l, normalize=True)
    lm.fit(x, y)
    crossvalscore = cross_val_score(lm, x, y, cv=10, scoring=meanquadraticloss)
    #df2 = pd.DataFrame([[l, crossvalscore.mean()]], columns=['Alpha','Ridge MSE 10CV'])
    arr.loc[i] = [l, crossvalscore.mean()]
    i = i + 1
```

# 3 Naive Bayes (10pt)

### 1. Read in the data (canvas-files-house votes). Ensure you know it's structure.

```python
data = pd.read_table('house-votes-84.csv.bz2', compression = "bz2", sep=",", header=None)
```

### 2. Compute these probabilities.

```python
temp = data.drop([0], axis=1)
temp.apply(pd.value_counts)
data.apply(pd.value_counts).T

democrats = data.apply(pd.value_counts).T['democrat'][0]
republicans = data.apply(pd.value_counts).T['republican'][0]
print('Pr(Republicans): ', republicans/ (democrats + republicans) * 100, '%')
print('Pr(Democrats): ', democrats/ (democrats + republicans) * 100, '%')
no = data[data[1] == "n"]
yes = data[data[1] == "y"]
```

```python
arr = pd.DataFrame(columns = ["Pr(voted no on issue i|democrat)", "Pr(voted yes on issue i|republican)", "Pr(voted no on issue i|republican)", "Pr(voted yes on issue i|democrat)"])
j = 0
for i in data.columns:
    if (i != 0):
        no = data[data[i] == 'n']
        yes = data[data[i] == 'y']
        nodemocrat = len(no[no[0] == 'democrat']) / democrats
        yesrepublican = len(yes[yes[0] == 'republican'])/republicans
        noRepublican = len(no[no[0] == 'republican'])/republicans
        yesDemoncrat = len(yes[yes[0] == 'democrat'])/democrats
        print("Column ", i)
        print("Pr(voted no on issue i|democrat): ", '{:.01%}'.format(nodemocrat))
        print("Pr(voted yes on issue i|republican): ", '{:.01%}'.format(yesrepublican))
        print("Pr(voted no on issue i|republican): ", '{:.01%}'.format(noRepublican))
        print("Pr(voted yes on issue i|democrat): ", '{:.01%}'.format(yesDemoncrat))
        print()
        df2 = pd.DataFrame([[nodemocrat, yesrepublican, noRepublican, yesDemoncrat]], columns=["Pr(voted no on issue i|democrat)", "Pr(voted yes on issue i|republican)", "Pr(voted no on issue i|republican)", "Pr(voted yes on issue i|democrat)"])
        arr.loc[j] = [nodemocrat, yesrepublican, noRepublican, yesDemoncrat]
        j = j + 1
arr
```

### 3. create prediction of being democrat and being republican for each representative based just on the prior.
### Note: if you operate with valid probabilities, obviously Pr(D) = 1 - Pr(R). However, this is not the case if you ignore the normalizer and compute in logarithms.

```python
print('Pr(Republicans): ', republicans/ (democrats + republicans) * 100 , '%')
print('Pr(Democrats): ', democrats/ (democrats + republicans) * 100, '%')
```

## 4. Now update the information, vote by vote: for each vote take the corresponding conditional proba-bility, and update the Bayes estimate accordingly.

```python
result = pd.DataFrame(columns = ["Demo vs Repu"])
index = 0
repu = republicans / (democrats + republicans)
demo = democrats / (democrats + republicans)

for i in range(len(data)):
    temp = data.loc[i]
    repon = repu
    demon = demo
    k = 0
    for j in range(len(data.columns) - 1):
        if (j != 0):
            vote = temp[j]
            if vote == 'n':
                repon = repon * arr['Pr(voted no on issue i|republican)'][k]
                demon = demon * arr['Pr(voted no on issue i|democrat)'][k]
            elif vote == 'y':
                repon = repon * arr['Pr(voted yes on issue i|republican)'][k]
                demon = demon * arr['Pr(voted yes on issue i|democrat)'][k]
        k = k + 1
    if repon > demon:
        result.loc[index] = "republican"
    else:
        result.loc[index] = "democrat"
    index = index + 1        
result
```

## 5. After including information about all 16 votes, you have your prediction. Print the confusion matrix, and calculate accuracy, precision and recall.

```python
from sklearn.metrics import confusion_matrix
print("Confusion Matrix:")
df = pd.DataFrame(confusion_matrix(data[0], result['Demo vs Repu']))
df.rename(index={0: "Actual: Democrat", 1:  "Actual: Republican"}, columns={0: "Predicted: Democrat", 1: "Predicted: Republican"})
```
