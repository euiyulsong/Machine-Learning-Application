```python
# 2 How Often Do We Get Big Outliers?
# 2.1 The Easy Task: Normal Sample

# Pick your number of repetitions R and sample size N. 1000 and 100 are good choices
r = 1000
n = 100
```

```python
# Now generate a sample of N independent standard normal random variables, and and it's mean.
# It's almost never exactly 0. How big is it in your case?
normal = np.random.normal(size = n)
mean = np.mean(normal)

print("Mean is ", np.mean(normal))
```

```python
# 3. Compute the 95% condence intervals for your results. Does the value you got fall into 95% confidence region?

confidence_interval = (np.percentile(normal, 5), np.percentile(normal, 95))
print("Confidence interval:")
print(confidence_interval)
if (confidence_interval[0] > mean):
    print("mean is smaller than 95% confidence interval")
elif (confidence_interval[1] < mean):
    print("mean is larger than 95% confidence interval")
else:
    print("mean fall into 95% confidence interval")
```

```python
# 4. Print mean-of-means (mean of m-s) and it's empirical 95% confidence region based on your experiment
array = np.zeros(r)
for value in range(0, r):
    normal = np.random.normal(size = 100)
    mean = np.mean(normal)
    array[value] = mean

print("Mean-of-mean =", np.mean(array))
print("Empirical 95% Confidence Region =", [np.percentile(array, 5), np.percentile(array, 95)])
```

```python
# 5. Plot the histogram of your m-s and mark this range on it.
ax = plt.axes(frameon=False)
result = plt.hist(array, rwidth=1000, color="green", bins=50)
plt.title("Histogram of M-s", fontsize='20')
plt.xlabel('Mean Value', fontsize='14')
plt.ylabel('Frequency', fontsize='14')
```
