```python
print("Use the example matrix A and plot it's image using a few different transformation.")

def rotation_matrix(A, alpha):
    alpha = np.radians(alpha)
    return A.dot(np.matrix([[np.cos(alpha), -np.sin(alpha)], [np.sin(alpha), np.cos(alpha)]]))
B = np.matrix([[0, 0], [0, 1], [0.5, 0.5]])

plt.plot(rotation_matrix(B, 0))
plt.plot(rotation_matrix(B, 45))
plt.plot(rotation_matrix(B, 90))
plt.plot(rotation_matrix(B, -225))
print("")
```

```python
print("Create a new simple object of your choice. The object should be asymmetric. Print this matrix.")
temp = np.matrix([[1, 2], [4, 8]])
print(temp)
```

```python
print("Show the object (rotation 0) and a few rotated images of that object.")
plt.plot(rotation_matrix(temp, 0))
plt.plot(rotation_matrix(temp, 45))
plt.plot(rotation_matrix(temp, 90))
print("")
```

```python
print("Construct the ip-x matrix Fx. This matrix should mirror (ip) the x components but leave the y components untouched. Demonstrate this with your own object.")
A = np.flip(temp, 0)
print(A)
print("")
```

```python
print("Create the stretch-y matrix Sy(s). This matrix should stretch (or compress) the y components by amount s while leaving x components untouched. Demonstrate it with your object.")
def stretchy_y(A, num):
    return A.dot(np.matrix([[1, 0], [0, num]]))

print(stretchy_y(temp, 2))
print("")
```
